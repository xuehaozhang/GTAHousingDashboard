```python
def process_bedrooms(value):
    value_str = str(value)
    if '+' in value_str:
        parts = value_str.split('+')
        total_bedrooms = sum(int(part) for part in parts)
        return total_bedrooms
    elif value_str.isnumeric():  
        return int(value_str)
    else:
        return value


gta_sale['Building.Bedrooms'] = gta_sale['Building.Bedrooms'].apply(process_bedrooms)
```
